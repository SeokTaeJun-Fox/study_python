## 1. 리스트 컬렉션 메서드

```python
li1 = [i + 1 for i in range(10)]
```

```python
# 추가
li1.append(10)
```

```python
li1
```

```python
for i in range(10, 21):
    li1.append(i)
```

```python
li1
```

```python
# list의 모든 합을 구하기
# min(): 가장 작은 값을 찾아서 반환하는 함수
# max(): 가장 큰 값을 찾아서 반환하는 함수
# sum(): 모든 값을 더하는 함수
sum(li1)
sum = 0
for i in li1:
    sum+=i
print(sum)
```

```python
li2 = ['a', 'b', 'c']
li3 = [10, 20, 30]
# 새로운 배열
li2 + li3
```

```python
# 원본 배열에 추가
li2.extend(li3)
```

```python
li2
```

```python
# 2번 인덱스에 100추가
li3.insert(2, 100)
```

```python
li3
```

```python
li3.sort()
```

```python
li3.sort(reverse=True)
```

```python
# .remove
li3.remove(10)
```

```python
li2.pop(2)
```
