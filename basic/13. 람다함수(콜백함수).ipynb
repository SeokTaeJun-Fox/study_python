```python
# 두 수를 더하고 2배 곱한 결과를 출력
def add(num1, num2, callback):
    callback(num1 + num2)

def get_square(num, callback):
    callback(num * 2)
```

```python
# lambda
add(10, 20, lambda result: get_square(result, print))
```

```python
# 성과 이름을 전달하면 풀 네임으로 붙이고 "님"을 붙여서 반환
def get_pull_name(last_name, first_name, callback):
    callback(last_name + first_name)

def add_nim(full_name, callback):
    callback(full_name + "님")

get_pull_name("홍", "길동", lambda full_name: add_nim(full_name, print))
```

## map, filter

```python
li1 = [(i + 1) * 10 for i in range(10)]
li1
```

```python
def multiple_2(num):
    return num * 2;

# map(함수, 객체)
list(map(multiple_2, li1))

# 람다함수는 값을 리턴
list(map(lambda i: i * 2, li1))
```

```python
# 65 => 90
li2 = [i for i in range(65, 91)]

print(li2)
```

```python
li3 = list(map(lambda c: chr(c), li2))
print(li3)
```

```python
# li3 소문자로 변경 li4를 생성

li4 = list(map(lambda c: c.lower(), li3))
print(li4)
```

```python
# li2로 aBcDeFgHiJ...로 변경 후 li5를 생성
def change(num):
    c = chr(num)
    c = c.lower() if num % 2 == 1 else c.upper()
    return c

li5 = list(map(change, li2))
print("".join(li5))
```

```python
li6 = [i + 1 for i in range(100)]
print(li6)
```

```python
# 3의 배수
li7 = list(filter(lambda n: n % 3 == 0, li6))
print(li7)
```

```python
# li6에 3의 배수들만 걸러내고 2배로 값을 바꾸기
li8 = list(map(lambda n: n * 2, filter(lambda n: n % 3 == 0, li6)))
print(li8)
```

```python
# li4에서 a ~ g까지만 filter로 걸러내고 li9 만들기
li9 = list(filter(lambda c: ord(c) > 103, li4))
print(li9)
```
