```python
lotto = [i for i in range(1, 46)]
print(lotto)
```

```python
# 한개 랜덤으로 뽑아오기
print(random.choice(lotto))
```

```python
# 무작위로 6개 뽑아오기
print(random.sample(lotto, 6))
```
