```python
# 이름(name), 나이(age), 운동종목(sports)
# 플레이어들을 소개하는 메서드

# 운동선수 3명의 객체화 하고
# 3명을 각각 소개하는 메서드를 사용해서 정보들을 출력하세요.
# ex) 000선수의 나이 00살 운동종목 0000
class Player:
    def __init__(self, name, age, sports):
        self.name = name;
        self.age = age;
        self.sports = sports;

    def introduce(self):
        print(f"{self.name}선수의 나이 {self.age}살 운동종목 {self.sports}")

p1 = Player("URA", 20, "축구")
p2 = Player("DaNon", 30, "농구")
p3 = Player("Sarry", 25, "야구")
players = [p1, p2, p3]

for player in players:
    print(f"이름 : {player.name} 나이 : {player.age} 종목 : {player.sports}")
```

```python
# 완전 객체 지향형 언어
# Market
# 판매
class Market:
    def __init__(self, market_name, product_name, product_price):
        self.market_name = market_name;
        self.product_name = product_name
        self.product_price = product_price

    def sell(self, customer):
        customer.money -= self.product_price * (100 - customer.discount) / 100

# Customer
# 할인율
class Customer:
    def __init__(self, name, money, discount):
        self.name = name
        self.money = money
        self.discount = discount
```

```python
cu = Market("CU", "도시락", 10000)
customer = Customer("이규혁", 50000, 10)

cu.sell(customer)
print(customer.money)
```

```python
# Atm, User 객체
# deposit

# Atm.deposit(User, 10000)
# 유저의 돈은 10000 차감,
# 은행에 돈은 10000 입금

# 은행이 가진돈과 유저의 돈 출력

class Atm:
    def __init__(self, money):
        self.money = money

    def deposit(self, user, money):
        if user.money >= money:
            user.money -= money
            self.money += money
        else:
            print("잔액이 부족합니다")

class User:
    def __init__(self, money):
        self.money = money

player = User(100000)
atm = Atm(100000)

atm.deposit(player, 10000)

print(player.money, atm.money)
```

```python
# 클래스 메서드
class Korean:
    # 클래스 변수
    country = "한국"

    # 클래스 메서드
    # 인스턴스X, 클래스변수O
    @classmethod # 데코레이터
    def trip(cls, country):
        if cls.country == country:
            print("국내여행")
        else:
            print("해외여행")

    # 정적 메서드
    @staticmethod
    def slogan():
        print("Imagine your korea")
```

```python
hong = Korean()

hong.trip("한국")
hong.slogan()

Korean.trip("미국")
Korean.slogan()
```

## 상속 (inheritance)

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age
```

```python
hong = Person("홍길동")
jang = Student("장보고", 20)

# 부모 => 타입도 같이 전달!
# 모든 자식은 부모 타입이다.
# 부모는 자식타입이 아니다.
print(isinstance(hong, Person))
print(isinstance(hong, Student))
print(isinstance(jang, Person))
print(isinstance(jang, Student))
```

```python
class Animal:
    def __init__(self, name, feed, species):
        self.name = name
        self.feed = feed
        self.species = species

    def feeding(self):
        print(f"{self.name}은 먹이를 먹습니다.")
        

class Dog(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)
    
    def act(self):
        print(f"{self.name}은 산책을 합니다.")

class Cat(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)

    def sleep(self):
        print(f"{self.name}은 잠을 자주 잡니다.")
```

```python
dog = Dog("누렁이", "잡식", "강아지")
dog.act()
dog.feeding()

cat = Cat("야옹이", "츄르", "고양이")
cat.sleep()
cat.feeding()
```

```python
# 사람(부모), 학생(자식)
# 사람 - 이름, 나이, 키, 몸무게
# 학생 - 학생, 핸드폰

# 사람은 인사하는 메서드
# 학생은 공부하는 메서드
```

```python
class Person:
    def __init__(self, name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

    def greeting(self):
        print(f"{self.name}는 인사를 합니다")

class Student(Person):
    def __init__(self, name, age, height, weight, stu_id, phone):
        super().__init__(name, age, height, weight)
        self.stu_id = stu_id
        self.phone = phone

    def study(self):
        print(f"{self.name}는 공부를 합니다")
```

```python
p1 = Person("URA", 30, 175, 75)
p1.greeting()

s1 = Student("DaNon", 30, 175, 75, "2013222111", "010-1234-5678")
s1.study()
```

## 다형성

```python
class Car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def start_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 켭니다")

    def stop_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 끕니다")

class SuperCar(Car):
    def start_engine(self):
        print(f"{self.brand}는 음성으로 시동을 켭니다")

    def stop_engine(self):
        print(f"{self.brand}는 음성으로 시동을 끕니다")
```

```python
matiz = Car("마티즈", 1000000)
matiz.start_engine()
matiz.stop_engine()
```

```python
lamborghini = SuperCar("람보르기니", 500000000)
lamborghini.start_engine()
lamborghini.stop_engine()
```
