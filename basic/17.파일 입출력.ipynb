## with문
- 사용한 리소스를 자동으로 닫아주는 문법(관리해주는 문법)

```python
with open("./food_list.txt", "w", encoding="utf-8") as f:
    foods = ["돈까스", "제육볶음", "김치찌개", "순대국밥", "오징어볶음"]

    for food in foods:
        f.write(food + "\n")

    print("food_list.txt가 작성되었습니다.")
```

```python
with open("./food_list.txt", "r", encoding="utf-8") as f:
    # print(f.read())
    foods = f.readlines()
    print(foods)

    # foods.index("김치찌개\n")

    if "김치찌개\n" in foods:
        foods[foods.index("김치찌개\n")] = "순두부찌개\n"

    print(foods)
# with open("./food_list.txt", "w", encoding="utf-8") as f:
#     f.writelines(foods)

with open("./food_list.txt", "w", encoding="utf-8") as f:
    foods = "".join(foods)
    f.write(foods)
```

## CSV(Comma Separated Value)

```python
import csv
with open("./test_csv.csv", "w", encoding="utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(("학번", "이름", "전화번호"))
    writer.writerows([(1, "홍길동", "01012341234"), (2, "장보고", "01012341234")])
```
