```python
class User:
    name: str
    age: int
    address: str
    phone: str

    def __init__(self, name, age, address, phone):
        self.name = name
        self.age = age
        self.address = address
        self.phone = phone

    def print_hobby(self, hobby: str) -> None:
        print(f"{self.name}의 취미는 {hobby}입니다.")
```

```python
jang = User("장보고", 30, "서울시 종로구", "010-4567-4567")

jang.name
```

```python
from pydantic import BaseModel
```

```python
class User(BaseModel):
    name: str
    address: str
    hobby: str
```

```python
# 키가 더 추가되면 틀 이외에 값은 무시
# 타입이 다를 때 오류 검증
user2_data: User = {
    "name": "장보고",
    "address": "경기도 구리시",
    "hobby": "코딩"
}
```
