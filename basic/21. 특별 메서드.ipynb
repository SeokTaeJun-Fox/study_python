```python
class User:
    name: str
    age: int
    phone: str
    
    def __init__(self, name, age, phone):
        self.name = name
        self.age = age
        self.phone = phone

    # toString
    def __str__(self):
        return f"User(name: {self.name}, age: {self.age}, phone: {self.phone})"
```

```python
print(User("홍길동", 20, "010-1234-1234"))
```

```python
from pydantic import BaseModel
```

```python
class Animal(BaseModel):
    name: str
    age: int
    feed: str
    hobby: str
    master: str
    gender: str
```

```python
cat: Animal = {
    "name": "고양이",
    "age": 2,
    "feed": "잡식",
    "hobby": "숙면",
    "master": "자신",
    "gender": "남"
}
```

```python
Animal(**cat)
```

## 연산자

```python
# 회사원
class Employee(BaseModel):
    id: int
    name: str
    income: int
```

```python
employee_data1 = {
    "id": 1,
    "name": "홍길동",
    "income": 10000
}

employee_data2 = {
    "id": 2,
    "name": "이순신",
    "income": 20000
}

employee1 = Employee(**employee_data1)
employee2 = Employee(**employee_data2)
```

```python
print(employee1)
print(employee2)
```

```python
# 학생
class Student(BaseModel):
    id: int
    name: str
    income: int
```

```python
student_data1 = {
    "id": 1,
    "name": "김철수",
    "income": 10000
}

student_data2 = {
    "id": 2,
    "name": "김영희",
    "income": 20000
}

student1 = Student(**student_data1)
student2 = Student(**student_data2)
```

```python
print(employee1)
print(employee2)
print(student1)
print(student2)
```

```python
print(Employee.__eq__(employee1, employee1))
print(Employee.__eq__(employee1, employee2))
print(Employee.__eq__(employee1, student1))
print(Employee.__eq__(employee1, student2))
```

```python
student1 = Student(**student_data1)
student3 = Student(**student_data1)
```

```python
print(student1)
print(student3)
```

```python
print(Student.__eq__(student1, student3))
```

```python
# 문제 생길수 있는 코드
class Student:
    def __init__(self, id, name):
        self.id = id
        self.name = name

    def __eq__(self, other):
        # 해시 주소
        if self.__repr__() == other.__repr__():
            return True

        # 타입
        if isinstance(other, Student):
            # id 끼리 값 비교
            return self.id == other.id

        return False
```

```python
student4 = Student(1, "홍길동")
student5 = Student(1, "홍길동")

Student.__eq__(student4, student5)
```
