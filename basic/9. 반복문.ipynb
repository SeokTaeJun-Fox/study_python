```python
# 일반 for 문
# 빠른 for 문
data1 = "apple" + "banana"

# 순차적으로 순회
# 객체의 개수만큼 반복
for i in data1:
    print(i)
```

## 컴프리헨션

```python
[i * 2 for i in [10, 20, 30]]
```

```python
{i * 2 for i in [10, 20, 30]}
```

```python
name = "장보고", "홍길동", "이순신"
```

```python
name[0]
```

```python
for i in name:
    print(i)
```

```python
# 인덱스를 만들던가
# enumerate: 인덱스를 만드는 함수
#[(0, 장보고), (1, 홍길동), (2, 이순신)]
# unpacking
for index, name in enumerate(name):
    print(f"{index + 1}번 {name}")

```

```python
# 일반 for문(range)
# range(10) => ~ 10
# range(10, 20) => 10 ~ 10
# range(20, 30, 2) => 20 ~ 30, 2개씩

"a"[0:]
"a"[10:20]
"a"[10:20:2]

for i in range(10):
    print(i)
    
```

```python
# 1 ~ 50 까지 홀수만 출력하기
for i in range(50):
    if i % 2 == 1:
        print(i)
```

```python
# 1 ~ 50까지 홀수만 리스트에 담아서 출력
# 컴프리헨션 문법 사용
odd = [i for i in range(50) if i % 2]
print(odd)
```

```python
set = {i for i in range(50) if i <= 5}
print(set)
```

```python
foods = {
    "a" : "떡갈비",
    "b" : "라면",
    "c" : "떡볶이",
    "d" : "불고기"
}
```

```python
# index 접근
for i in foods:
    print(foods.get(i))
```

```python
foods.values()
```

```python
for i in list(foods.values()):
    print(i)
```

```python
for key, food in foods.items():
    print(key, food)
```

## while 문

```python
count = 0
while count < 5:
    print("반복문")
    count += 1
    
```

```python
count = 0

while (count := count + 1) < 5:
    print("반복문")
```
