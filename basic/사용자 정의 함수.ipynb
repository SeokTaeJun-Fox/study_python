## 값을 몇개나 전달할 지 모를 때

```python
# 값을 몇 개를 전달해도 모두 더하는 함수
# args: 가변인자, 가변매개변수, rest parameter

def total(*args):
    print(args)
    print(type(args))
    return sum(args)
```

```python
total(10, 20)
```

```python
# 두 수를 전달하면 사칙연산의 결과 반환
def calc(num1, num2):
    return num1 + num2, num1 - num2, num1 * num2, num1 / num2
```

```python
add, sub, mul, div = calc(10, 20)
```

```python
print(add)
print(sub)
print(mul)
print(div)
```

## keywards argument

```python
def order_info(*args, name, address, phone):
    total = sum(args)
    print(f"{name}님의 주소 {address}, 연락처: {phone}")
    print(f"총 구매 가격: {total}")
```

```python
# 이름, 주소, 연락처, 총 구매가격
order_info(1000, 50000, 15000, 7000, name = "홍길동", address="강남", phone="010123456789")
```

```python
user_info = {
    "name": "홍길동",
    "address" : "강남",
    "phone": "01012341234"
}
```

```python
user_info
```

```python
order_info(1000, 50000, 15000, 7000, **user_info)
```
