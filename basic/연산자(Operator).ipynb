## 8. 삼항 연산자

```python
bool("True" if 11 == 10 else "False")
```

```python
# 삼항 두 수 비교
# 삼항 연산자 엔터치면 안됨.
# 들여쓰기로 영역을 구분한다.
num1 = input("첫 번째 정수\n")
num2 = input("두 번째 정수\n")

print(num1 if num1 > num2 else "두 수가 같습니다" if num1 == num2 else num2)
```

```python
num1 = 20
print(num1)
```

```python
# 사용자에게 정수 또는 실수를 입력 받아서
# 정수인지 실수인지 확인 후 삼항 연산자로 결과를 출력하시오.
# ex) 10 -> "정수"
# ex) 11.0 -> "정수"
# ex) 11.3 -> "실수"

# message

num1 = input("숫자를 입력하시오")
check = float(num1) - float(num1) // 1
print("정수" if check == 0 else "실수")
```
