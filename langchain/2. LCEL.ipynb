## temperature(창의력)
- 0.0 ~ 0.3: 결정론 - 요약, 추출, 일관됨, 정확성
- 0.5 ~ 0.7: 균형적 - 적당히 자연스러움 유연성
- 0.8 ~ 1.0+: 무작위 - 창위적이고 예측 불가능

```python
class NewLineOutputParser(BaseOutputParser):
    # 반드시 parse
    def parse(self, text):
        lines = text.split("\n")
        return [line.lstrip("-123456789. ").strip() for line in lines]
```

```python
newline_parser = NewLineOutputParser()
```

```python
newline_parser.parse("""- 1. 햄버거 \n- 2. 떡볶이\n- 3. 치킨""")
```

```python
template = ChatPromptTemplate.from_messages([
  ("system", """
        리스트를 생성하는 기계입니다.
        요청한 모든 리스트에 개수는 최대 {max_length}개 까지만 목록으로 표시하세요.
        그 이상 초과되는 리스트는 답변하지 마세요.
  """),
    ("human", "{question}")
])

prompt = template.format_messages(
    max_length=5,
    question="AI를 잘하려면 어떤 것부터 공부해야해?"
)
```

## 랭체인(Chain 생성)
- "|": 파이프 연산자로 체인을 만든다

```python
# list
first_chain = template | chat | NewLineOutputParser()
```

```python
chain_result = first_chain.invoke({
    "max_length": 5,
    "question": "AI를 잘하려면 어떤 것부터 공부해야해?"
})
```

```python
chain_result
```

```python
# RunnableSequence
# 이전 단계의 출력이 다음 단계의 입력으로 자동 전달되는 파이프라인 객체
print(type(first_chain))
```

## 1. 템플릿 생성

```python
template = ChatPromptTemplate.from_messages([
  ("system", """
      당신은 세계적인 수준의 여행 가이드입니다.
      사람들이 좋아하는 여행 장소를 많이 알고 있습니다.
      설명없이 지역 명소 이름만 목록으로 {max_length}개 까지 답변하세요.
      목록의 개수가 초과하는 것을 답변하지 마세요.
  """),
  ("human", """
      {place} 여행 장소 추천해줘!
  """)
])
```

```python
second_chain = template | chat
```

```python
trip_result = second_chain.invoke({
    "max_length": 3,
    "place": "부산"
})
```

```python
print(trip_result.content)
```

```python
# 저녁 메뉴 추천(양식, 중식, 한식 선택할 수 있도록)
# 체인을 생성 후 결과를 출력
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", """
        당신은 세계적인 요리사입니다.
        {food_type}의 음식 메뉴 추천해줄 수 있습니다.
    """),
    ("human", "{question}"),
])

food_chain = prompt | chat
```

```python
food_result = food_chain.invoke({
    "food_type": "한식",
    "question": "오늘 저녁을 어떤 것을 먹어야 잘 먹었다고 소문날까?"
})
```

```python
food_result.content
```

## .stream()
- 결과를 실시간 chunk 단위로 쪼개서 응답한다.

```python
food_result = food_chain.stream({
    "food_type": "한식",
    "question": "오늘 저녁을 어떤 것을 먹어야 잘 먹었다고 소문날까?"
})
```

```python
for chunk in food_result:
    print(chunk.content, end="", flush=True)
```

```python
# 형태로 딕셔너리
# 최근 급등한 주식 리스트 5개
```

```python
template = ChatPromptTemplate.from_messages([
    ("system", """
        넌 주식의 트레이더야!
        사용자가 주식과 관련된 질문을 하면 {max_length}개의 주식을 추천해줘

        반드시 아래의 JSON 형식으로 답변해줘
        출력 형식:
        {{
            "one": "tesla",
            "two": "samsung",
            ...
            "five": "apple"
        }}
    """),
    ("human", "{question}"),
])
```

```python
stock_chain = template | chat
```

```python
stock_result = stock_chain.invoke({
    "max_length": 5,
    "question": "최근 급등한 미국 주식 추천해줘!"
})
```

```python
stock_result.content
```

```python
import json

```

```python
json.loads(stock_result.content)
```

```python
class JsonOutputParser(BaseOutputParser):
    def parse(self, text):
        return json.loads(text)
```

```python
chat = ChatOpenAI(temperature=0)

stock_chain = template | chat | JsonOutputParser()
```

```python
stock_result = stock_chain.invoke({
    "max_length": 5,
    "question": "최근 급등한 미국 주식 추천해줘!"
})
```
