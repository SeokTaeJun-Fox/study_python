## 2단계: 예시용 프롬포트 템플릿 정의

```python
example_template = """
    Human: {country}
    AI: {answer}
"""

example_prompt = PromptTemplate.from_template(example_template)
example_prompt
```

## 3단계: FewShotPromptTemplate 생성 후 결합

```python
prompt = FewShotPromptTemplate(
    # 질문
    example_prompt = example_prompt,
    
    # 예시
    examples=examples,

    # 사용자의 질문
    suffix="Human: {country}에 대해서 어떻게 알고 있어요?",
    input_variables=["country"]
)
prompt
```

```python
prompt
```

```python
chat = ChatOpenAI(temperature=0)
```

```python
chain = prompt | chat
```

```python
result = chain.invoke({
 "country": "한국"
})
```

## Chat 모델 (LengthBasedExampleSelector)

```python
examples = [
    { "country": "프랑스에 대해서 어떻게 알고 있나요?", 
      # 원하는 형식의 답변 
      "answer": """ 
      저는 이렇게 알고 있어요. 
      수도: 파리 언어: 
      프랑스어 음식: 와인과 치즈 
      통화: 유로 
      """ },
    {
        "country": "일본에 대해서 어떻게 알고 있나요?",
        "answer": """
            저는 이렇게 알고 있어요.
            수도: 도쿄
            언어: 일본어
            음식: 스시와 라멘
            통화: 엔
        """
    },
    {
        "country": "이탈리아에 대해서 어떻게 알고 있나요?",
        "answer": """
            저는 이렇게 알고 있어요.
            수도: 로마
            언어: 이탈리아어
            음식: 피자와 파스타
            통화: 유로
        """
    },
    {
        "country": "브라질에 대해서 어떻게 알고 있나요?",
        "answer": """
            저는 이렇게 알고 있어요.
            수도: 브라질리아
            언어: 포르투갈어
            음식: 페이조아다와 슈하스코
            통화: 헤알
        """
    }
]
```

```python
length_prompt = PromptTemplate(
    input_variables=["country", "answer"],
    template="""
        Human: {country}에 대해서 어떻게 알고 있어요? \n
        AI: {answer}
    """
)

example_prompt = ChatPromptTemplate.from_messages([
    ("human", "{country}에 대해서 어떻게 알고 있어요?"),
    ("ai", "{answer}"),
])

example_selector = LengthBasedExampleSelector(
    examples=examples,
    example_prompt=example_prompt,
    max_length=200
)


fewshot_prompt = FewShotChatMessagePromptTemplate(
    example_selector=example_selector,
    example_prompt=example_prompt
)
```

```python
final_prompt = ChatPromptTemplate.from_messages([
    ("system", "당신은 지리학 전문가입니다."),
    fewshot_prompt,
    ("human", "{country}에 대해 어떻게 알고 있어요?"),
])
```

```python
chain = final_prompt | chat
```

```python
result = chain.invoke({
    "country": "대한민국"
})
```

```python
print(result.content)
```
