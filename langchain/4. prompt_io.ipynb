```python
from dotenv import load_dotenv
import os

load_dotenv()
print(os.environ.get('OPENAI_API_KEY')[:20])
```

```python
base_path = r"D:\back_0900_stj\python\resource"
```

```python
base_path
```

```python
file_path = os.path.join(base_path, "prompt.json")
```

```python
file_path
```

```python
with open(file_path, "w") as f:
    f.write("""
        {
            "_type": "prompt",
            "template": "{country}에 대해서 어떻게 알고 있어요?",
            "input_variables": ["country"]
        }
    """)
```

```python
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import load_prompt
```

```python
chat = ChatOpenAI(temperature=0.1)
```

```python
prompt = load_prompt(file_path)
```

```python
chain = prompt | chat
```

```python
result = chain.invoke({
    "country": "한국"
})
```

```python
result.content
```
