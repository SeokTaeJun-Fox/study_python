## LCEL 방법 권장

```python
# MessagesPlaceholder: 프롬프트 안에 “대화 기록”을 끼워 넣는 자리표시 역할
# InMemoryChatMessageHistory: 대화 내용을 메모리에 저장
# RunnableWithMessageHistory: 체인 + 메모리 + 프롬프트를 자동으로 연결해주는 랩퍼
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain_core.chat_history import InMemoryChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
```

```python
chat = ChatOpenAI(temperature=0.1)
```

```python
prompt = ChatPromptTemplate.from_messages([
    ("system", """
        휴먼을 도와주는 AI 도구입니다. 친절하게 답변하세요.
    """),
    # 저장소 메모리 위치
    MessagesPlaceholder(variable_name="chat_history"),
    ("human", "{question}"),
])
```

```python
chain = prompt | chat
```

# 저장소(메모리) 생성

```python
# redis, db
store = {}

def get_history(session_id):
    if session_id not in store:
        store[session_id] = InMemoryChatMessageHistory()
    return store[session_id]
```

```python
chain_with_memory = RunnableWithMessageHistory(
    chain,
    get_history,
    input_messages_key="question",
    history_messages_key="chat_history"
)
```

```python
result = chain_with_memory.invoke(
    {
        "question": "내 이름은 DaNon이야!",
    },
    config={"configurable": {"session_id": "user1"}}
)
```

```python
print(result)
```

```python
store
```

```python
print(store["user1"])
```

```python
result = chain_with_memory.invoke(
    {
        "question": "난 오늘 저녁에 김치찌개를 먹을거야!",
    },
    config={"configurable": {"session_id": "user1"}}
)
```

```python
result = chain_with_memory.invoke(
    {
        "question": "난 오늘 저녁에 무엇을 먹을 것 같아?",
    },
    config={"configurable": {"session_id": "user1"}}
)
```

```python
print(store["user1"])
```

```python
chain_with_memory = RunnableWithMessageHistory(
    chain,
    get_history,
    input_messages_key="question",
    history_messages_key="chat_history"
).with_config(
    config={"session_id": "user2"}
)
```

```python
result = chain_with_memory.invoke({
    "question": "난 오늘 저녁에 무엇을 먹을 것 같아?",
})
```
