## TikToken=
- length_function=
- OpenAI의 Chat 계열 모델이 텍스트를 처리하는 데 사용하는 토큰 단위(token)을 계산하고 변환하는 라이브러리
- splitter의 len()가 length_fucntion의 default로 사용된다. 또한 CharacterTextSplitter에서는 "글자의 개수"를 chunk로 카운트한다.
- 하지만 OpenAI에서는 token의 개수를 세므로 다르다. 따라서 맞춰줄 필요가 있다.

### 4. 프롬프트 작성

```python
# 최초 프롬프트
initial_prompt = ChatPromptTemplate.from_template("""
    제공된 context에서만 질문에 답하세요.
    context:
        {context}

    Answer the question:
        {question}
""")

# refine 프롬프트
refine_prompt = ChatPromptTemplate.from_template("""
    당신의 임무는 최종 요약본을 작성하는 것입니다.
    지금까지의 요약된 내용이 있습니다.
    exsisting_answer:
        {exsisting_answer}

    이제 요약을 개선하는 데 도움이 될 수 있는 추가 context가 있습니다.
    context:
        {context}

    새로운 context를 사용해서 원본을 요약하고 다듬어 
    질문에 대한 답변만 하세요.
    새로운 context에서 유용한 정보가 없다면, 원래 요약을 그대로 반환하세요.
""")

# refine_prompt

# # 최초 프롬프트
# initial_prompt = ChatPromptTemplate.from_template("""
#     Given there following context:
#     {context}

#     Answer the question:
#     {question}
# """)

# # refine 프롬프트
# # refine_prompt = ChatPromptTemplate.from_template("""
# #     당신의 임무는 최종 요약본을 작성하는 것입니다.
# #     지금까지의 요약된 내용이 있습니다.
# #     exsisting_answer:
# #         {exsisting_answer}

# #     이제 요약을 개선하는 데 도움이 될 수 있는 추가 context가 있습니다.
# #     context:
# #         {context}

# #     새로운 context을 사용해서 원본을 요약하고 다듬습니다.
# #     새로운 context에서 유용한 정보가 없다면, 원래 요약을 그대로 반환하세요.
# # """)
# refine_prompt = ChatPromptTemplate.from_template("""
#     You are refining an existing summary with newly retrieved context.
    
#     Current summary:
#     {existing_answer}
    
#     New context:
#     {context}
    
#     Instructions:
#     - Update and improve the current summary using only relevant information from the new context.
#     - Preserve important existing information unless contradicted by the new context.
#     - Keep the summary concise, coherent, and non-redundant.
#     - If the new context does not provide meaningful additional information, return the original summary unchanged.
#     - Do not invent facts or include unsupported assumptions.
    
#     Refined summary:
# """)
```

```python
chat = ChatOpenAI(
    temperature=0.5
)

def refine_docs(inputs):
    docs, question = inputs["docs"], inputs["question"]

    # 첫 번째 문서로 시작
    chain = initial_prompt | chat
    result = chain.invoke({
        "context": docs[0].page_content,
        "question": question
    }).content

    # 두 번째 문서부터 refine 순회
    refine_chain = refine_prompt | chat
    for doc in docs[1:]:
        result = refine_chain.invoke({
            "exsisting_answer": result,
            "context": doc.page_content,
            "question": question
        }).content

    return result
```

```python
refine_chain = {
    "docs": retrieval,
    "question": RunnablePassthrough()
} | RunnableLambda(refine_docs)
```

### Map re-rank documents chain

```python
import re
```

```python
class MapReRankChain:
    def __init__(self, llm, retriever):
        self.llm = llm
        self.retriever = retriever

        # 1. 문서의 관련성 평가하는 프롬프트
        self.re_rank_prompt = ChatPromptTemplate.from_template("""
            다음 context의 질문과 얼마나 관련있는지
            0-10까지의 척도 내에서 숫자로만 대답하세요.
    
            context: 
                {context}
                
            question: 
                {question}
        """)
    
        self.final_prompt = ChatPromptTemplate.from_template("""
            아래 context에서 가장 관련성이 높은 정보를 바탕으로 question에 답하세요.
    
            context:
                {context}
                
            question:
                {question}
        """)
    
        self.re_rank_chain = self.re_rank_prompt | self.llm | StrOutputParser()
        self.final_chain = self.final_prompt | self.llm | StrOutputParser()

    # invoke 메서드 생성
    def invoke(self, question: str):
        docs = self.retriever.invoke(question)

        # re-rank 문서의 점수
        scored_docs = []
        for doc in docs:
            score_str = self.re_rank_chain.invoke({"context": doc.page_content, "question": question})

            # 숫자만 추출
            score = float(re.search(r"\d+", score_str).group()) if re.search(r"\d+", score_str) else 0
            scored_docs.append((score, doc))

        # 정렬 상위 k개
        scored_docs.sort(key=lambda tu: tu[0], reverse=True)
        top_docs = [doc for score, doc in scored_docs[:2]]
        
        print(f"재정렬된 docs: {scored_docs}")

        # 최종 답변 생성
        context = "\n\n".join(doc.page_content for doc in top_docs)
        return self.final_chain.invoke({"context": context, "question": question})
```

```python
re_rank_chain = MapReRankChain(llm=llm, retriever=retriever)
```

```python
result = re_rank_chain.invoke("제비가 물어온 것은?")
```
